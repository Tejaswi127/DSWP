# WEIGHT CONVERSION

```python
from tkinter import *
root=Tk()
def conversion():
    gram = float(val.get())*1000
    pound = float(val.get())*2.2046
    ounce = float(val.get())*35.274
    t1.insert(END,gram)
    t2.insert(END,pound)
    t3.insert(END,ounce)
root.geometry("500x100")
root.title("weight conversion")
l=Label(root,text="enter your weight")
val=StringVar()
e2=Entry(root,textvariable=val)
e3 = Label(root, text = 'Gram')
e4 = Label(root, text = 'Pounds')
e5 = Label(root, text = 'Ounce')
t1 = Text(root, height = 1, width = 20)
t2 = Text(root, height = 1, width = 20)
t3 = Text(root, height = 1, width = 20)

b1 = Button(root, text = "Convert", command = conversion)
l.grid(row = 0, column = 0)
e2.grid(row = 0, column = 1)
e3.grid(row = 2, column = 0)
e4.grid(row = 3, column = 0)
e5.grid(row = 4, column = 0)
t1.grid(row = 2, column = 1)
t2.grid(row = 3, column = 1)
t3.grid(row = 4, column = 1)
b1.grid(row = 1, column = 1)

root.mainloop()
```

# LOGIN 

```python
from tkinter import *
root=Tk()
def submit1():
    if(val1.get()=='teja' and val2.get()=='velpula'):
            t1.insert(END,"succesfully logged in")
    else:
            t1.insert(END,"wrong credentials")
root.geometry("500x100")
root.title("Login")
l1=Label(root,text="username")
val1=StringVar()
e1=Entry(root,textvariable=val1)
l2=Label(root,text ="password")
val2=StringVar()
e2=Entry(root,textvariable=val2)
b1 = Button(root, text = "submit", command =submit1)
t1 = Text(root, height = 1, width = 20)
l1.grid(row = 0, column = 0)
l2.grid(row = 1, column = 0)
e1.grid(row = 0, column = 1)
e2.grid(row = 1, column = 1)
t1.grid(row=3,column=1)
b1.grid(row = 2, column = 1)
root.mainloop()
```

# FLAMES 

```python
from tkinter import *
root=Tk()
def flames():
    n1=val1.get()
    n2=val2.get()
    a=n1
    b=n2
    for i in range(len(n1)):
        for j in range(len(n2)):
            if(n1[i]==n2[j]):
                a=a.replace(n1[i]," ")
                b=b.replace(n1[i],"")
    l=['f','l','a','m','e','s']
    x=len(a)+len(b)
    while(len(l)>1):
        index=(x%len(l))-1
        if(index>=0):
            right=l[index+1:]
            left=l[:index]
            l=right+left
        else:
            l=l[:len(l)-1]
    t1.delete("1.0",END)
    t1.insert(END,l)
root.geometry("500x100")
root.title("Flames")
l1=Label(root,text="name1")
val1=StringVar()
e1=Entry(root,textvariable=val1)
l2=Label(root,text ="name2")
val2=StringVar()
e2=Entry(root,textvariable=val2)
b1 = Button(root, text = "submit", command=flames)
t1 = Text(root, height = 1, width = 20)
l1.grid(row = 0, column = 0)
l2.grid(row = 1, column = 0)
e1.grid(row = 0, column = 1)
e2.grid(row = 1, column = 1)
t1.grid(row=3,column=1)
b1.grid(row = 2, column = 1)
root.mainloop()
```
